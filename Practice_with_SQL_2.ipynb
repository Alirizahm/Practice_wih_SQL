```python
import pandas as pd
import pandasql

from pandasql import sqldf 
sql_run = lambda q: sqldf(q, globals())
```

```python
worldcity = pd.read_csv('worldcities.csv', sep = ';', encoding = 'latin-1')
uscity = pd.read_csv('uscities.csv', sep = ';', encoding = 'latin-1')
uscity.columns = ['city', 'state_id', 'state_name', 'county_fips', 'county_name',
       'lat', 'lng', 'population', 'density', 'source', 'military',
       'incorporated', 'timezone', 'ranking', 'zips', 'id']
```

## Question 1

Find cities from worldcity that meet all these conditions:

- The population of the city exceeds the population of the most population city in the Philippines. (Meaning: for example city X is the city in the Philippines with the most population. We want to list the cities with a population greater than city X), AND
- There is city in United States (uscity) has same name as the selected cities

Show only cities has name that meet two condition above, no need another columns.

NOTE: Remove duplicate rows.

Do only 1x query, don't do more than 1 query

```python
query = """
SELECT DISTINCT city
FROM worldcity
WHERE worldcity.population > (SELECT population FROM worldcity WHERE country = 'Philippines')
AND city IN (SELECT city FROM uscity)
"""
sql_run(query)  
```

```python
query = """
SELECT DISTINCT a.city
FROM worldcity a
INNER JOIN uscity b
On a.city = b.city
WHERE a.population > (SELECT population FROM worldcity WHERE country = 'Philippines')
"""
sql_run(query)  
```

## Question 2

Find cities in United States (uscity) that meet all these conditions:

- Has the name 'city' which is not the same as 'state_name', and also
- Has a population above the average population of the cities in 'county_name' Miami-Dade, and also
- Consists of two or more words (example: Los Angeles, New York)
- The expected output is only the 'city' column, no need for other columns

Do only 1x query, don't do more than 1 query

```python
query = """
SELECT city
FROM uscity 
WHERE population > (SELECT AVG(population) FROM uscity WHERE county_name = 'Miami-Dade')
AND city NOT IN (state_name)
AND city like '% %'
"""
sql_run(query)   
```
